```python
import pandas as pd

# Load the dataset
file_path = '/content/globalterrorismdb_0718dist.csv'  # Update if needed
terrorism_data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Basic information about the dataset
terrorism_data.info()

# Display the first few rows
terrorism_data.head()
```

```python
# Check for missing values
missing_values = terrorism_data.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the number of attacks per year
plt.figure(figsize=(10,6))
sns.countplot(x='iyear', data=terrorism_data, palette='viridis')
plt.xticks(rotation=90)
plt.title('Number of Terrorist Attacks per Year')
plt.show()

# Top 10 countries with the most attacks
top_countries = terrorism_data['country_txt'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='inferno')
plt.title('Top 10 Countries with the Most Terrorist Attacks')
plt.xticks(rotation=90)
plt.show()
```

```python
# Hot zones by region
hot_zones_region = terrorism_data['region_txt'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=hot_zones_region.index, y=hot_zones_region.values, palette='coolwarm')
plt.title('Regions with the Most Terrorist Attacks')
plt.xticks(rotation=90)
plt.show()
```

```python
# Common attack types
attack_types = terrorism_data['attacktype1_txt'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=attack_types.index, y=attack_types.values, palette='magma')
plt.title('Most Common Types of Terrorist Attacks')
plt.xticks(rotation=90)
plt.show()
```
